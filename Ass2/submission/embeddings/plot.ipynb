```python
import gensim
from sklearn.decomposition import TruncatedSVD
from matplotlib import pyplot as plt
import numpy as np
# Plot the following words or other words you are interested in
# You better pick those words that look different in the 2D space compared with the LSA vectors
words = ['学', '习', '曰', '子', '人', '仁']
words_pinyin = ['xue', 'xi', 'yue', 'zi', 'ren1', 'ren2']
```

```python
embedding_file = "/home/stu_12310401/nlp/SUSTech-NLP25/Ass2/emb50_k2_win1_embeddings.txt"
model = gensim.models.KeyedVectors.load_word2vec_format(embedding_file, binary=False)
print(model.key_to_index)
selected_embeddings = np.array([model[word] for word in words])
# print(selected_embeddings)
svd1 = TruncatedSVD(n_components=2, random_state=42)
M1 = svd1.fit_transform(selected_embeddings)
print(M1)
plt.figure(figsize=(6, 6))
for i, word in enumerate(words):
    plt.scatter(M1[i, 0], M1[i, 1])
    plt.annotate(words_pinyin[i], (M1[i, 0], M1[i, 1]))
plt.title('emb50_k2_win1')
plt.savefig('results/emb50_k2_win1.png')
plt.show()
```

```python
from PIL import Image
import os

# 获取results文件夹中的所有图片文件
image_folder = 'results'
image_files = [f for f in os.listdir(image_folder) if f.endswith('.png')and f.startswith('emb')]

# 加载所有图片
image_files.sort()
images = [Image.open(os.path.join(image_folder, file)) for file in image_files]

# 获取单张图片的宽度和高度
width, height = images[0].size

# 创建一个新的空白图片，宽度为4张图片宽度之和，高度为2张图片高度之和
combined_width = width * 4
combined_height = height * 2
combined_image = Image.new('RGB', (combined_width, combined_height))

# 将每张图片粘贴到新图片上
for i, image in enumerate(images):
    x = (i % 4) * width
    y = (i // 4) * height
    combined_image.paste(image, (x, y))

# 保存拼接后的图片
# combined_image.save('results/combined_image_4x2.png')
combined_image.show()
```
